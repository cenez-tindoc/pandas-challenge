```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
#Total number of players
#Calculate the total number of players by using SN as it is the unique value
player=len(purchase_df["SN"].value_counts())

#Create summary data frame
player_total=pd.DataFrame([player], columns = ["Total Players"])

player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate the reqs
#No. of Unique Items
unique_total=len(purchase_df["Item ID"].unique())

#Average Price
average=round(purchase_df["Price"].mean(),2)

#No. of Purchases
purchases=len(purchase_df["Purchase ID"].unique())

#Total Revenue
revenue=purchase_df["Price"].sum()

#Create summary data frame
#Use the calculated data and create the columns with labels
analysis=pd.DataFrame({"Number of Unique Items":[unique_total], "Average Price":[average], "Number of Purchases":[purchases], "Total Revenue":[revenue]})
prices=['Average Price', 'Total Revenue']

#Format the columns with prices using $X.XX format 
analysis[prices]=analysis[prices].applymap(lambda x:'${:.2f}'.format(x))

analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create a new df to remove duplicated SN
gender=purchase_df[["SN","Gender"]]

#Remove the duplicates in the list
gender=gender.drop_duplicates()

#Count the players per gender
gender_count=gender["Gender"].value_counts()

#Getting the data for the demographics table
values=[gender_count[0], gender_count[1], gender_count[2]]
percentages=[(gender_count[0]/player)*100,(gender_count[1]/player)*100,(gender_count[2]/player)*100]

#Creating the DF
#Use the calculated data and create the columns and indeces with labels
demo_table = pd.DataFrame({"Total Count":values,"Percentage of Players":percentages})
demo_table.index=(["Male", "Female", "Other / Non-Disclosed"])

#Format the columns with prices using $X.XX format 
demo_table['Percentage of Players']=demo_table['Percentage of Players'].map('{:.2f}%'.format)

demo_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Create the data
#Using groupby Gender and calculating the reqs
group=purchase_df.groupby(["Gender"])
pur_count=group["SN"].count()
pur_avg=group["Price"].mean()
pur_tot=group["Price"].sum()
pur_person=pur_tot/pur_count

#Creating the DF
#Use the calculated data and create the columns with labels
pur_gender = pd.DataFrame({"Purchase Count": pur_count, "Average Purchase Price": pur_avg,"Total Purchase Value": pur_tot,
                              "Avg Total Purchase per Person": pur_person})
gender_prices=['Average Purchase Price', 'Total Purchase Value','Avg Total Purchase per Person']

#Format the columns with prices using $X.XX format 
pur_gender[gender_prices]=pur_gender[gender_prices].applymap(lambda x:'${:.2f}'.format(x))

pur_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create an new df to remove duplicated SN but using Age data
nondup_df=purchase_df[["SN","Age"]]

#Remove the duplicates
nondup_df=nondup_df.drop_duplicates()

#Create bins per age group
bins=[0,9,14,19,24,29,34,39,200]

#Create labels for the bins
grp_lbl=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Slice the data and place them into each bin
pd.cut(nondup_df["Age"],bins,labels=grp_lbl).head()

#Place the data series into a new column inside of the DF
nondup_df["Age Ranges"]=pd.cut(nondup_df["Age"],bins,labels=grp_lbl)

#Create a Groupby object based on "Age Ranges"
age_range=nondup_df.groupby("Age Ranges")

#Find how many players in each range/bin and its percentages
tot_count=age_range["Age"].count()
percentage=tot_count/player*100

#Creating the DF
#Use the calculated data and create the columns with labels
agedemo_df=pd.DataFrame({"Total Count":tot_count, "Percentage of Players":percentage})

#Format the columns with prices using $X.XX format 
agedemo_df['Percentage of Players']=agedemo_df['Percentage of Players'].map('{:.2f}%'.format)

agedemo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a Groupby object based on SN
spender=purchase_df.groupby("SN")

#Calculate the reqs
sn_purct=spender["Purchase ID"].count()
sn_prcavg=spender["Price"].mean()
sn_purtot=spender["Price"].sum()

#Create the DF
#Create new columns and use the calculated reqs data
spender_df=pd.DataFrame({"Purchase Count":sn_purct,"Average Purchase Price":sn_prcavg,"Total Purchase Value":sn_purtot})

#Sort with from the highest Total Purchase Value
spender_df=spender_df.sort_values("Total Purchase Value", ascending=False)

#Format the columns with prices using $X.XX format and label the columns
spender_prices=['Average Purchase Price', 'Total Purchase Value']
spender_df[spender_prices]=spender_df[spender_prices].applymap(lambda x:'${:.2f}'.format(x))

spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a Groupby object
popular=purchase_df.groupby(['Item ID', 'Item Name'])['Price'].agg(['count','mean','sum'])
popular.columns=['Purchase Count','Average Purchase Price','Total Purchase Value']

#Sort according to descending Purchase Count
popular_df=popular.sort_values("Purchase Count", ascending=False)

#Format the columns with prices using $X.XX format and label the columns
pop_prices=['Average Purchase Price', 'Total Purchase Value']
popular_df[pop_prices]=popular_df[pop_prices].applymap(lambda x:'${:.2f}'.format(x))

popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Use the same dataframe "popular"
profit_df=popular.sort_values('Total Purchase Value',ascending=False)

#Format the columns with prices using $X.XX format and label the columns
profit_df[pop_prices]=profit_df[pop_prices].applymap(lambda x:'${:.2f}'.format(x))

profit_df.head()
```
